```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def hilbert(n):
    x = np.arange(1, n+1) + np.arange(0, n)[:, np.newaxis]
    return 1.0/x

def swapRow(A,r1,r2):
    # put row r2 in r1 spot
    print('Swapping rows')
    A_new = np.copy(A)
    for i in range(len(A_new)):
        temp_var = A_new[r1][i]
        A_new[r1][i] = A_new[r2][i]
        A_new[r2][i] = temp_var

    return A_new





def gaussianElimination(A,b): #nxn matrices only
    A_temp = np.copy(A)
    A_inv = np.identity(len(A))
    b_temp = np.copy(b)

    #making zeros below diagnal
    for c in range(len(A)): #columns
        if(A_temp[c][c] == 0): # zero along the diagnal
                for i in range(c,len(A[0])):
                    if(A_temp[i][c] != 0):
                        A_temp = np.copy(swapRow(A_temp,c,i))
                        A_inv = np.copy(swapRow(A_inv,c,i))
        for r in range(c+1,len(A[0])): #rows
            #print('r: '+str(r)+'        c: '+str(c)+'        d: '+str(c)+','+str(c))
            if(A_temp[r][c] != 0):
                temp = A_temp[r][c]*(-1.0)/A_temp[c][c] # value @ index trying to get to zero *(-1)/value along diagnal
                b_temp[r][0] = b_temp[r][0] + temp*b_temp[c][0]
                for i in range(c,len(A)):
                    A_temp[r][i] = A_temp[r][i] + temp*A_temp[c][i]
                for i in range(len(A)):
                    A_inv[r][i] = A_inv[r][i] + temp*A_inv[c][i]

    
    #print('Above time')
    #print()
    #making zeros above diagnal
    for c in range(1,len(A)): #columns
        if(A_temp[c][c] == 0): # zero along the diagnal
                for i in range(c,len(A[0])):
                    if(A_temp[i][c] != 0):
                        A_temp = np.copy(swapRow(A_temp,c,i))
                        A_inv = np.copy(swapRow(A_inv,c,i))
        for r in reversed(range(c)): #rows
            #print('r: '+str(r)+'        c: '+str(c)+'        d: '+str(c)+','+str(c))
            if(A_temp[r][c] != 0):
                temp = A_temp[r][c]*(-1.0)/A_temp[c][c] # value @ index trying to get to zero *(-1)/value along diagnal
                b_temp[r][0] = b_temp[r][0] + temp*b_temp[c][0]
                for i in range(c,len(A)):
                    A_temp[r][i] = A_temp[r][i] + temp*A_temp[c][i]
                for i in range(len(A)):
                    A_inv[r][i] = A_inv[r][i] + temp*A_inv[c][i]


    #print('A: '+str(A_temp))
    #print('A_inv: '+str(A_inv))
    #print('b: '+str(b_temp))

    #print('Make diagnals 1')
    #print()
    #making diagnals 1
    for i in range(len(A)):
        temp = 1.0/A_temp[i][i]
        b_temp[i][0] *= temp
        A_temp[i][i] *= temp
        for j in range(len(A)):
            A_inv[i][j] *= temp





    #print('The solution is correct if: '+str(np.matmul(A_inv,b) == b_temp))
    #print('AA^-1 = I -->'+str(np.matmul(A,A_inv) == np.identity(len(A))))

    print('The solution is correct if A^-1b == b_')


    return A_inv,b_temp,A_temp


```

```python
m = 4
a = hilbert(m)
#b = np.array([[0.564348], [0.93798], [0.707275],[0.0248771]])
b = np.array([[1], [2], [3],[4]])

t = gaussianElimination(a,b)
print('My programs inverse to problem #9 was:')
print(t[0])
print('The solution to the system of linear equations in problem #9 is:')
print(t[1])
print()

```
